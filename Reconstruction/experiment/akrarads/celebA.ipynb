```python
import glob
import numpy as np
import shutil
```

```python
imgs = glob.glob('./dataset/tigers/*')
```

```python
print(len(imgs))
idx = np.arange(0,len(imgs))
np.random.shuffle(idx)
```

```python
len(idx[:1000])
```

```python
desc = './tigers/'
count = 21
for index in idx[:1000-20]:
    src_path = imgs[index]
    file_ext = src_path.split('.')[2]
    desc_path = f'{desc}{count}.{file_ext}'
    print(src_path,desc_path)
    shutil.copy(src_path, desc_path)
    count = count + 1
    # break
```
